```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv")
df.head()
```

```python
df2 = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv")
df2.head()
```

```python
df3 = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv")
df3.head()
```

```python
df.columns = [column.lower().replace(" ", "_") for column in df.columns]
df2.columns = [column.lower().replace(" ", "_") for column in df2.columns]
df3.columns = [column.lower().replace(" ", "_") for column in df3.columns]
```

```python
df['customer_lifetime_value'] = df['customer_lifetime_value'].str.replace('%', '').astype(float)
```

```python
df2['customer_lifetime_value'] = df2['customer_lifetime_value'].str.replace('%', '').astype(float)
```

```python
df = df.rename(columns={'st': 'state'})
```

```python
df2 = df.rename(columns= {'st' : 'state'})
```

```python
df.drop_duplicates(inplace=True)
df.isna().sum()
```

```python
df2.isna().sum()
```

```python
df2.drop_duplicates(inplace=True)
df2.isna().sum()
```

```python
df3.drop_duplicates(inplace=True)
df3.isna().sum()
```

```python
concated_df = pd.concat([df, df2, df3], ignore_index=True)
concated_df.sample
```

```python
concated_df.sample(10)
```

```python
concated_df.replace({'1/0/00': '1'}, inplace=True)
concated_df
```

```python
concated_df['total_claim_amount'] = concated_df['total_claim_amount'].round(2)
```
